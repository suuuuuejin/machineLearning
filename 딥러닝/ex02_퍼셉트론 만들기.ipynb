```python
import numpy as np
```

```python
# 활성화 함수(step function):
# 0기준으로 이상이면1, 미만 0
def activation_func(x):  # x는 선형함수에서 y값
  if x <=0:
    return 0
  else:
    return 1
```

#### AND 퍼셉트론 만들기

```python
def AND(x1,x2):
  w1=1
  w2=1
  b= -1.5
  y= w1*x1 + w2*x2 +b
  return activation_func(y)
```

```python
print(AND(1,1))
print(AND(1,0))
print(AND(0,1))
print(AND(0,0))
```

#### OR퍼셉트론 만들기

```python
def OR(x1,x2):
  w1=1
  w2=1
  b=0
  y=w1*x1 + w2*x2 +b
  return activation_func(y)
```

```python
print(OR(1,1))
print(OR(1,0))
print(OR(0,1))
print(OR(0,0))
```

#### NAND 퍼셉트론 만들기

```python
def NAND(x1,x2):
  w1= -1
  w2= -1
  b= 1.5
  y= w1*x1 + w2*x2 +b
  return activation_func(y)
```

```python
print(NAND(1,1))
print(NAND(1,0))
print(NAND(0,1))
print(NAND(0,0))
```

#### XOR MLP 만들기

```python
def XOR(x1,x2):
  a1= NAND(x1,x2)
  a2= OR(x1,x2)
  return AND(a1,a2)
```

```python
print(XOR(1,1))
print(XOR(1,0))
print(XOR(0,1))
print(XOR(0,0))
```
